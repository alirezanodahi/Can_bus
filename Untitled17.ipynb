```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.metrics import confusion_matrix, precision_recall_curve, roc_curve, auc
from sklearn.ensemble import IsolationForest
import itertools
from collections import Counter
# important library

%matplotlib inline
# read the dataset
df4 = pd.read_csv('/content/drive/MyDrive/datasets/ML_350/ML350_Dataset.csv',sep=',')
#drop flag and ready to train
df5 = df4.drop(['Flag'], axis=1)


model=IsolationForest(n_estimators=70, max_samples=0.6, contamination=float(0.5), max_features=0.8)
model.fit(df5)

df5['anomaly']=model.predict(df5)

df5= df5.join(df4['Flag'])

tp=len(df5[(df5['anomaly'] == -1 )& (df5['Flag'] == "T" )].reset_index())
fp=len(df5[(df5['anomaly'] == -1 )& (df5['Flag'] == "R" )].reset_index())
fn=len(df5[(df5['anomaly'] == 1 )& (df5['Flag'] == "T" )].reset_index())
tn=len(df5[(df5['anomaly'] == 1 )& (df5['Flag'] == "R" )].reset_index())
print('tp = ',tp,"fp = ",fp,"fn = ",fn,"tn = ",tn)
precision = tp / (tp + fp)

# Calculate recall
recall = tp / (tp + fn)

# Calculate F1 score
f1 = 2 * (precision * recall) / (precision + recall)

precision, recall, f1
accuracy = (tp+tn)/(tp+tn+fn+fp)
accuracy
```

```python
# Visualization of anomalies
plt.figure(figsize=(10, 6))
plt.scatter(df5.iloc[:, 0], df5.iloc[:, 1], c=df5['anomaly'], cmap='coolwarm', marker='x')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Anomalies Detected by Isolation Forest')
plt.colorbar(label='Anomaly Score')
plt.grid(True)
plt.show()
```

```python
# Confusion matrix
cm = confusion_matrix(df5['Flag'], np.where(df5['anomaly'] == -1, 'T', 'R'))
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['R', 'T'], yticklabels=['R', 'T'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Convert 'Flag' column to binary format
df5['Flag_binary'] = df5['Flag'].apply(lambda x: 1 if x == 'T' else 0)

# Precision-Recall curve
precision, recall, _ = precision_recall_curve(df5['Flag_binary'], np.where(df5['anomaly'] == -1, 1, 0))
plt.figure(figsize=(8, 6))
plt.step(recall, precision, color='b', alpha=0.2, where='post')
plt.fill_between(recall, precision, step='post', alpha=0.2, color='b')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('Precision-Recall Curve')
plt.show()
```

```python
# ROC curve
fpr, tpr, _ = roc_curve(df5['Flag_binary'], np.where(df5['anomaly'] == -1, 1, 0))
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
# Use a smaller subset of data
subset = df5.sample(frac=0.1, random_state=42)

# Train Isolation Forest on the smaller subset
model = IsolationForest(n_estimators=70, max_samples=0.6, contamination=0.5, max_features=0.8)
model.fit(subset.drop(['anomaly', 'Flag', 'Flag_binary'], axis=1))

# Visualize one of the trees in the Isolation Forest
plt.figure(figsize=(20, 10))
plot_tree(model.estimators_[0], feature_names=subset.columns[:-3], filled=True, rounded=True)
plt.title('Isolation Forest Tree (Smaller Subset)')
plt.show()
```
