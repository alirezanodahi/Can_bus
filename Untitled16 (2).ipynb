```python
tp=len(df5[(df5['anomaly'] == -1 )& (df5['Flag'] == "T" )].reset_index())
fp=len(df5[(df5['anomaly'] == -1 )& (df5['Flag'] == "R" )].reset_index())
fn=len(df5[(df5['anomaly'] == 1 )& (df5['Flag'] == "T" )].reset_index())
tn=len(df5[(df5['anomaly'] == 1 )& (df5['Flag'] == "R" )].reset_index())
print('tp = ',tp,"fp = ",fp,"fn = ",fn,"tn = ",tn)
precision = tp / (tp + fp)

# Calculate recall
recall = tp / (tp + fn)

# Calculate F1 score
f1 = 2 * (precision * recall) / (precision + recall)

precision, recall, f1
```
